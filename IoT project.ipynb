```python
#attributes = list(data_set_1.columns[:8])
#X = data_set_1[attributes].values 
#y = data_set_1['diabetes'].values

attributes = list(data_set_2.columns[:8])
X = data_set_2[attributes].values 
y = data_set_2['diabetes'].values
```

```python
from sklearn.preprocessing import StandardScaler 
regularizer = StandardScaler() 
X = regularizer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state =0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
from sklearn import model_selection
```

```python
classifiers = []
classifiers.append(('LR', LogisticRegression(solver='liblinear')))
classifiers.append(('KNN', KNeighborsClassifier()))
classifiers.append(('NB', GaussianNB()))
classifiers.append(('SVM', SVC(gamma='auto')))
classifiers.append(('DT', DecisionTreeRegressor()))
```

```python
classifier_results = []
classifier_names = []
```

```python
k_fold = model_selection.KFold(n_splits = 10, random_state = 6)
for classifier_name, classifier in classifiers:
    cv_score = model_selection.cross_val_score(classifier, X, y, cv = k_fold, scoring= 'accuracy')
    classifier_results.append(cv_score)
    classifier_names.append(classifier_name)
    result = "%s: %.4f +/- %.4f" % (classifier_name, cv_score.mean(), cv_score.std())
    print(result)
```

```python
fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(1,1,1)
plt.boxplot(classifier_results)
ax.set_xticklabels(classifier_names)
plt.savefig("comparison_2.pdf")
```

```python
from sklearn.model_selection import GridSearchCV

c_list = list(np.arange(1, 100))

param_grid_lr= [
    {
        'C': c_list,
        'penalty': ['l1'],
        'solver': ['liblinear'],
        'multi_class': ['ovr'],
        'random_state': [42]
    },
    {
        'C': c_list,
        'penalty': ['l2'],
        'solver': ['liblinear', 'newton-cg', 'lbfgs'],
        'multi_class': ['ovr'],
        'random_state': [42]
    }
]

optional_lr = GridSearchCV(LogisticRegression(), param_grid_lr, cv = k_fold, scoring='accuracy')

optional_lr.fit(X, y)

print(optional_lr.best_params_)
```

```python
lr_model = LogisticRegression(C = 1, multi_class = 'ovr', penalty = 'l2', solver = 'newton-cg',random_state = 42)
lr_model.fit(X_train, y_train)

lr_model_predict = lr_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, classification_report

print("Confusion matrix:")
print("{0}\n".format(confusion_matrix(y_test, lr_model_predict, labels=[1,0])))

print('Accuracy: {:.2f}%\n'.format(accuracy_score(y_test, lr_model_predict) * 100))

print('Classification report:\n', classification_report(y_test, lr_model_predict))
print('Training set score: {:.2f}%'.format(lr_model.score(X_train, y_train) * 100))
print('Testing set score: {:.2f}%'.format(lr_model.score(X_test, y_test) * 100))
```

```python
lr_model_predict_prob = lr_model.predict_proba(X_test)[:, 1]
fp, tp, thresholds = roc_curve(y_test, lr_model_predict_prob)

plt.subplots(figsize=(12,10))
plt.plot([0,1],[0,1],'k--')
plt.plot(fp, tp)
plt.xlim([-0.01, 1.0])
plt.ylim([0.0, 1.01])
plt.savefig("roc_lr_2.pdf")
```

```python
param_grid_svm = {
    'C': [1.0, 10.0, 50.0],
    'kernel': ['linear', 'poly', 'sigmoid'],
    'shrinking': [True, False],
    'gamma': ['auto', 1, 0.1],
    'coef0': [0.0, 0.1, 0.5]
}

optional_svm = GridSearchCV(SVC(), param_grid_svm, cv = k_fold, scoring='accuracy')

optional_svm.fit(X, y)

print(optional_svm.best_params_)
```

```python
k_values = list(range(1,50))

neighbors = filter(lambda x: x % 2 != 0, k_values)

cv_scores = []
graph_x = []

for k in neighbors:
    graph_x.append(k)
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=k_fold, scoring='accuracy')
    cv_scores.append(scores.mean())
```

```python
MSE = [1 - x for x in cv_scores]

plt.subplots(figsize=(12,10))
plt.plot(graph_x, MSE)
```
